```python
import pandas as pd
import os
```

```python
file_names = os.listdir('../../data/spotify')
file_names
```

```python
df = pd.DataFrame()
for file_name in file_names:
    df_current = pd.read_csv('../../data/spotify/' + file_name)
    print(file_name, df_current.shape)
    df = pd.concat([df, df_current], ignore_index=True)
```
