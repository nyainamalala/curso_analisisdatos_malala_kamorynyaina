## bide

```python
import numpy as np
import urllib
```

```python
# https:archive-beta.ics.uci.edu/dataset/53/iris

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
response = urllib.request.urlopen(url)

data = np.genfromtxt(response, delimiter=",", usecols=(0, 1, 2, 3)) # de golpe todo
data[:10]
```

```python
print(data.ndim)
print(data.shape)
```

```python
print("Sepal length")
print(f"sepal length mean: {np.mean(data[:, 0]):.2f}")
print(f"sepal length median: {np.median(data[:, 0]):.2f}")
print(f"sepal length max: {np.median(data[:, 0]):.2f}")
print(f"sepal length min: {np.median(data[:, 0]):.2f}")
print(f"sepal length std: {np.std(data[:, 0]):.2f}")
```

```python
# sepal width in cm
print("Sepal width")
print(f"sepal width mean original: {np.mean(data[:, 1]):.2f}")
print(f"sepal width mean rounded: np.{np.mean(data[:, 1]):.2f}")

print(f"sepal width median: {np.mean(data[:, 1]):.2f}")
print(f"sepal width max: {np.max(data[:, 1]):.2f}")
print(f"sepal width min: {np.min(data[:, 1]):.2f}")
print(f"sepal width std: {np.std(data[:, 1]):.2f}")
```

```python
# Petal width
print("petal width")
print(f"petal width mean original: {np.mean(data[:, 3]):.2f}")
print(f"petal width mean rounded: np.{np.mean(data[:, 3]):.2f}")

print(f"petal width median: {np.mean(data[:, 3]):.2f}")
print(f"petal width max: {np.max(data[:, 3]):.2f}")
print(f"petal width min: {np.min(data[:, 3]):.2f}")
print(f"petal width std: {np.std(data[:, 3]):.2f}")
```

```python
# Petal length
print("Petal length")
print(f"petal length mean: {np.mean(data[:, 2]):.2f}")
print(f"petal length median: {np.median(data[:, 2]):.2f}")
print(f"petal length max: {np.median(data[:, 2]):.2f}")
print(f"petal length min: {np.median(data[:, 2]):.2f}")
print(f"petal length std: {np.std(data[:, 2]):.2f}")
```

```python
# Encontrar el indice de la flor con la longitud del pétalo más larga

PETAL_LENGTH = 2 # position

max_petal_length_index = np.argmax(data[:, PETAL_LENGTH])

max_petal_length_index = data[max_petal_length_index, PETAL_LENGTH]

print("Indice de la longitud del pétalo más larga", max_petal_length_index)
print("longitud del pétalo más larga", max_petal_length_index)
```
