```python
numeric_columns = ['Sales', 'Quantity', 'Discount', 'Profit', 'Shipping Cost']

for column in numeric_columns:
    plt.figure(figsize=(6,4))
    sns.histplot(data=df, x=column)
    plt.title(f'Hisplot de {column}')
    
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR # limite inferior
    upper_limit = Q3 + 1.5 * IQR # limite superior
    plt.xlim(lower_limit, upper_limit) #para no mostrarlo en la grafica
    
    plt.show()
```

# Analisis bivariante

## barplot: Sales / Profit totales por a√±o

2011 - 2015

```python
# Crear columna Year
df = pd.read_csv('../../../data/superstore_dataset2011-2015.csv')
# Pandas 2.0.1
# df['Order Date'] = pd.to_datetime(df['Order Date'], format='mixed')
# Pandas 1.5
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['year'] = df['Order Date'].dt.year

# groupby de Sales por Year
sales_per_year = df.groupby('year')['Sales'].sum() # serie
profit_per_year = df.groupby('year')['Profit'].sum() # serie

# barplot
sns.barplot(x=sales_per_year.index, y=sales_per_year.values)
```

## barplot de sales_per_year y profit_per_year

```python
# Crear un DataFrame de sales_per_year y mostrar una gr√°fica de √°rea
df_sales_per_year = pd.DataFrame({
    'Sales': sales_per_year.values, # serie
    'Profit': profit_per_year.values # serie
}, index=sales_per_year.index) # √≠ndice

df_sales_per_year.plot(kind='bar', alpha=0.5)
plt.title('Ventas y Beneficios por a√±o')
plt.ticklabel_format(style='plain', axis='y')  # Desactivar la notaci√≥n cient√≠fica en el eje Y
```
