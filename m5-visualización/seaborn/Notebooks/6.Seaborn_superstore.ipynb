```python
# Columnas num√©ricas
import numpy as np
df.select_dtypes(include=np.number).columns.to_list()
```

```python
numeric_columns = df.select_dtypes(include=np.number).columns.to_list()

for column in numeric_columns:
    plt.figure(figsize=(10,7))
    sns.histplot(data=df, x=column)
    plt.title(f'Hisplot de {column}')
    plt.show()
```

```python
numeric_columns = ['Sales', 'Quantity', 'Discount', 'Profit', 'Shipping Cost']

for column in numeric_columns:
    plt.figure(figsize=(6,4))
    sns.histplot(data=df, x=column)
    plt.title(f'Hisplot de {column}')
    plt.show()
```

```python
numeric_columns = ['Sales', 'Quantity', 'Discount', 'Profit', 'Shipping Cost']

for column in numeric_columns:
    plt.figure(figsize=(6,4))
    sns.histplot(data=df, x=column)
    plt.title(f'Hisplot de {column}')
    
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR # limite inferior
    upper_limit = Q3 + 1.5 * IQR # limite superior
    plt.xlim(lower_limit, upper_limit) #para no mostrarlo en la grafica
    
    plt.show()
```

```python
categorical_columns = df.select_dtypes(include='object').columns.to_list()

for column in categorical_columns:
    if df[column].nunique() < 10:
        plt.figure(figsize=(5, 6))
        sns.countplot(data=df, x=column)
        plt.title('Countplot de ' + column)
        plt.xticks(rotation=45)
        plt.show()
```

```python
date_columns = df.select_dtypes(include='datetime64').columns.to_list()

for column in date_columns:
     plt.figure(figsize=(14,7))
     sns.lineplot(data=df, x=column, y='Sales')
     plt.title('Ventas por ' + column)
     plt.show()
```

```python
date_columns = df.select_dtypes(include='datetime64').columns.to_list()

for column in date_columns:
     plt.figure(figsize=(14,7))
     
     # necesario index tipo fecha
     # df[column]['2011-01-01':'2012-01-01'].plot
     
     sns.lineplot(data=df, x=column, y='Sales')
     plt.title('Ventas por ' + column)
     plt.xlim(pd.Timestamp('2011-01-01'), pd.Timestamp('2012-01-01'))
     plt.show()
```

```python
date_columns = df.select_dtypes(include='datetime64').columns.to_list()

for column in date_columns:
     plt.figure(figsize=(14,7))
     
     # necesario index tipo fecha
     # df.set_index(column, inplace=True)
     # df[column]['2011-01-01':'2012-01-01'].plot
     
     sns.lineplot(data=df, x=column, y='Sales', errorbar=None)
     plt.title('Ventas por ' + column)
     plt.xlim(pd.Timestamp('2011-01-01'), pd.Timestamp('2012-02-01'))
     plt.ylim(0, 1000)
     plt.show()
```
